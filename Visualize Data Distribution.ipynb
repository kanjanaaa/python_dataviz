### scatter plot
#### การลงจุดแบบกระจัดกระจายบนกราฟ นิยมทั้ง 2 และ 3 มิติ

```python
irisdata = pandas.read_csv('C:\\Users\\Admin\\Downloads\\iris.data',header=None)
irisdata
```

##### ตัวหน้าหมายถึงแถว ตัวหลังหมายถึงหลัก(-3 ต้องการแค่ 2หลักหรือคอลัมล์)

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]  #ตารางแบบย่อ
```

##### ตั้งแต่ตัวแรก :

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1]) 
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)  ### ขยายขนาดจุดด้วย s
```

#### เลขหลักแรกแทน X เลขหลักสองแทน Y  เลขหลักสามแทน ขนาด

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'k') ### c แทนด้วยสี k = ดำ
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'c') ### c แทนด้วยสี c= ฟ้า r = แดง
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = colors)
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr')
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c =np.array([[1,0.5,0]])) ### สีแดงใส่ 1 สีเขียวใส่ 0.5 สีฟ้าใส่ 0 = สีส้ม
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c =np.array([[1,0.5,0.8]])) ### สีแดงใส่ 1 สีเขียวใส่ 0.5 สีฟ้าใส่ 0.8 = สีชมพู
```

```python
 plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr',marker="p") ###  marker = สัญลักษณ์
```
